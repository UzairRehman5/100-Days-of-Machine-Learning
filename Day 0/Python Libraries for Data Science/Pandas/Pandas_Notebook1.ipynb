 ## The value_counts() function
 
The value_counts() function is used to get a Series containing counts of unique values. The resulting object will be in descending order so that the first element is the most frequently-occurring element.

```python
# It will give total values of each unique index
df['winner'].value_counts()
```

```python
# It will give total values of each unique index
df['player_of_match'].value_counts().head()
```

```python
# It will give total values of each unique index
df['venue'].value_counts()
```

## The sort_values() method

1) Applicable on both series and dataframe
2) Inplace parameter will make permanent changes in dataframe, the default value of inplace is False

```python
# Sorting the values of series in ascending order
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values()
```

```python
# Sorting the values of series in descending order
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values(ascending = False)
```

```python
# Sorting the values of dataframe in ascending order
df.sort_values('city')
```

```python
# Sorting the values of dataframe in descending order
df.sort_values('city',ascending=False)
```

```python
# Inplace parameter will make permanent changes in dataframe, the default value of inplace is False
df.sort_values('date', inplace=True)
df
```

```python
# Sorting city in increasing order and after that date will be sorted in decreasing order
df.sort_values(['city', 'date'], ascending = (True, False))
```
