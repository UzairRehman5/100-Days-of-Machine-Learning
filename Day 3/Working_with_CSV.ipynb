# 3. Opening a csv file from an URL

If we need to load a CSV file located on a server, we can utilize this code snippet. In the snippet, all the code will remain unchanged; we simply need to insert the URL into the 'url' variable

```python
import requests
from io import StringIO

url = 'https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv'
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers = headers)
data = StringIO(req.text)

df2 = pd.read_csv(data)
df2
```

# 4. Sep Parameter

In CSV files, data is typically separated by commas, whereas in TSV files, data is separated by tabs instead of commas. This distinction in separators is crucial for proper data parsing. When working with TSV files, it's important to specify the 'sep' parameter as an additional step during the reading process. This parameter informs the parser about the specific delimiter used in the file, ensuring accurate data extraction.

We are also passing a list named 'names' that contains the names of all the columns.

```python
df3 = pd.read_csv('movie_titles_metadata.tsv', sep = '\t', names = ['Serial No', 'Movies_name', 'Release_Year', 'Rating', 'No_of_Upvotes', 'Genres'])
df3
```

# 5. Index_col parameter

If we wish to replace the default index generated by pandas with the index column of your dataset, we utilize this parameter.

```python
df4 = pd.read_csv("aug_train.csv", index_col = "enrollee_id")
df4
```

# 6. Header parameter

If the column names of a dataset have been transformed into rows and you need to revert them back to their original state, the 'header' parameter is employed

```python
df5 = pd.read_csv('test.csv',header=1)
df5
```

# 7. Use_cols parameter

When you have a dataset with many columns, and you are interested in working with only a subset of those columns, you can use the usecols parameter to indicate which specific columns you want to include in the resulting DataFrame. This can be useful for reducing memory usage and speeding up data loading and processing, especially when dealing with large datasets.

```python
df6 = pd.read_csv("aug_train.csv", usecols=['enrollee_id', 'enrolled_university', 'experience'])
df6
```

# 8. Squeeze parameters

The "squeeze" parameter is used to convert a DataFrame with a single column into a Series. 

```python
df7 = pd.read_csv("aug_train.csv", usecols=['education_level'], squeeze=True)
df7
```

# 9. Skip rows/ nrows parameter

The skip rows parameter is used to specify the number of rows to skip at the beginning of the file while reading data.

The nrow parameter is used to specify the number of rows to be read from a file when using functions like read.csv.

```python
df8 = pd.read_csv('aug_train.csv',skiprows=[1, 4])
df8
```

```python
df9 = pd.read_csv('aug_train.csv', nrows=10000)
df9
```

# 10. Error bad lines parameter

When reading data from files, especially in formats like CSV or TSV, there might be instances where a line does not conform to the expected structure. This could be due to various reasons such as data corruption, incomplete lines, or other formatting issues. In such cases, the 'bad lines' parameter provides options for how to handle these problematic lines.

```python
df10 = pd.read_csv("aug_train.csv", error_bad_lines=False)
df10
```

# 11. dtypes parameter

If we want to change the datatype of any column, we can easily change it while reading the file.

```python
df11= pd.read_csv('aug_train.csv',dtype={'target':int})
df11     
```

# 12. Handling Dates

The parse_dates parameter in Pandas is used to specify which columns in a CSV file should be parsed as dates.

```python
df12 = pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date'])
df12
```

# 13. Convertors

The converters parameter in pandas is used to pass a dictionary of functions to transform values on specified columns.

```python
def rename (name):
    if name == 'Kolkata Knight Riders':
        return 'KKR'
    
    else:
        return name
```

```python
df13 = pd.read_csv('IPL Matches 2008-2020.csv', converters={'team1':rename, 'team2':rename, 'toss_winner': rename, 'winner':rename})
df13
```

# 14. na_values parameter

The na_values parameter in pandas is used to specify a list of strings or regular expressions that should be treated as missing values.

```python
df14 = pd.read_csv('aug_train.csv', na_values=['never'])
df14
```

# 15. Loading a huge dataset in chunks

The chunksize parameter in pandas is used to read large datasets in chunks. This can be helpful for memory management and performance, especially when working with datasets that are too large to fit into memory all at once.

```python
df15 = pd.read_csv('aug_train.csv', chunksize=5000)
```

```python
# To access the chunk we created, we will take help of loop, for example if we want to check the shape of all the chunks, 
# we can do that easily with the help of loop

for chunks in df15:
    print(chunks.shape)
```
