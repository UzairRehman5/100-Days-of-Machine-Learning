# Importing Libraries

```python
# Import the NumPy library and alias it as 'np'
import numpy as np

# Import the Pandas library and alias it as 'pd'
import pandas as pd
```

# Importing Dataset

```python
# Read data from a CSV file named 'customer.csv' and store it in a Pandas DataFrame called 'df'
df = pd.read_csv('customer.csv')
```

```python
# Display the first few rows of the DataFrame 'df' to inspect the data
df.head()
```

# Data Preprocessing

```python
# Extracting features (X) and target variable (y) from the DataFrame 'df'.

# Extracting the first four columns of 'df' to create the feature matrix 'X'.
X = df.iloc[:, 0:4]

# Extracting the fifth/last column of 'df' to create the target variable 'y'.
y = df.iloc[:, 4]
```

```python
# Displaying the feature matrix 'X' which contains input values
X
```

```python
# Displaying the target variable 'y' which contains output values
y
```

# Label Encoding

```python
# Import the LabelEncoder class from scikit-learn's preprocessing module
from sklearn.preprocessing import LabelEncoder
```

```python
# Create an instance of the LabelEncoder class
le = LabelEncoder()
```

```python
# Fit the LabelEncoder to the target variable 'y' and transform 'y' into numerical labels
y = le.fit_transform(y)
```

```python
# Displaying the target variable 'y' which contains encoded output values
y
```

```python
# Access the 'classes_' attribute to retrieve the original categorical labels
# This attribute contains the unique categories mapped to numerical labels
encoded_classes = le.classes_
```
