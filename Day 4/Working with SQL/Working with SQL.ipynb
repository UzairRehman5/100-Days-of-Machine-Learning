# Establishing a connection to a MySQL database server

```python
# Establishing a connection to a MySQL database server using the given parameters.

conn = mysql.connector.connect(host='localhost', user='root', password='', database='World')
```

# Reading data from the database using the provided SQL query

#### 1) Read data from the 'city' table in the database using the provided SQL query

```python
df1 = pd.read_sql_query("SELECT * FROM city", conn)
df1
```

#### 2) Read data from the 'country' table in the database using the provided SQL query

```python
df2 = pd.read_sql_query("SELECT * FROM country", conn)
df2
```

#### 3) Read data from the 'countrylanguage' table in the database using the provided SQL query

```python
df3 = pd.read_sql_query("SELECT * FROM countrylanguage", conn)
df3
```

#### 4) Read data from the 'country' table where the 'Name' column is 'Pakistan'

```python
df4 = pd.read_sql_query("SELECT * FROM country WHERE Name LIKE 'Pakistan'", conn)
df4
```
